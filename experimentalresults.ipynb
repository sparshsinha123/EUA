## common deadline 10

```python
df_10 = pd.read_csv('res_10_3.csv').reset_index()
df_10.head()
```

```python
plt.figure(figsize = (15 , 7))
sns.set(style="ticks", rc={"lines.linewidth": 4})
sns.set_style(style="whitegrid")
sns.lineplot(x=df_10.index , y=df_10.lpp_user , label = "integer programming")
sns.scatterplot(x=df_10.index , y=df_10.lpp_user)

sns.lineplot(x=df_10.index , y=df_10.random_user, label = "random")
sns.scatterplot(x=df_10.index , y=df_10.random_user)

sns.lineplot(x=df_10.index , y=df_10.greedy_user, label = "greedy")
sns.scatterplot(x=df_10.index , y=df_10.greedy_user)

sns.lineplot(x=df_10.index , y=df_10.mcf_user, label = "most capacity first")
sns.scatterplot(x=df_10.index , y=df_10.mcf_user)

sns.lineplot(x=df_10.index , y=df_10.mflow_user, label = "constrained max flow")
sns.scatterplot(x=df_10.index , y=df_10.mflow_user)

plt.ylim(20,50)
plt.xlabel("Instances",fontsize=20)
plt.ylabel("Number of users allocated",fontsize=20)
plt.legend(fontsize=20)
# plt.title("number of users allocated for all the methods with common-deadline 10",fontsize=22)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.savefig('deadline_10_user.eps', format='eps')
plt.show()
```

```python
### the number of servers user data
data_frame = pd.read_csv('u_s_10_3.csv')
data_frame = data_frame.drop_duplicates(subset=['num_servers'],keep='first')
data_frame
```

```python
plt.figure(figsize = (15 , 7))
sns.set(style="ticks", rc={"lines.linewidth": 4})

sns.set_style(style="whitegrid")
sns.lineplot(x=data_frame.num_users , y=data_frame.lpp_user , label = "integer programming")
sns.scatterplot(x=data_frame.num_users , y=data_frame.lpp_user)

sns.lineplot(x=data_frame.num_users , y=data_frame.random_user, label = "random")
sns.scatterplot(x=data_frame.num_users , y=data_frame.random_user)

sns.lineplot(x=data_frame.num_users , y=data_frame.greedy_user, label = "greedy")
sns.scatterplot(x=data_frame.num_users , y=data_frame.greedy_user)

sns.lineplot(x=data_frame.num_users , y=data_frame.mcf_user, label = "most capacity first")
sns.scatterplot(x=data_frame.num_users , y=data_frame.mcf_user)

sns.lineplot(x=data_frame.num_users , y=data_frame.mflow_user, label = "constrained max flow")
sns.scatterplot(x=data_frame.num_users , y=data_frame.mflow_user)

plt.legend(fontsize=20)
plt.xlabel("Total number of users",fontsize=20)
plt.ylabel("Number of users allocated",fontsize=20)
# plt.title("Number of users allocated with common deadline 10",fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.savefig("allocated_users_vs_total_users.eps" , format='eps')

plt.show()
```

```python
plt.figure(figsize = (15 , 7))
sns.set(style="ticks", rc={"lines.linewidth": 4})

sns.set_style(style="whitegrid")
sns.lineplot(x=data_frame.num_users , y=data_frame.lpp_user / data_frame.num_users, label = "integer programming")
sns.scatterplot(x=data_frame.num_users , y=data_frame.lpp_user / data_frame.num_users)

sns.lineplot(x=data_frame.num_users , y=data_frame.random_user / data_frame.num_users, label = "random")
sns.scatterplot(x=data_frame.num_users , y=data_frame.random_user / data_frame.num_users )

sns.lineplot(x=data_frame.num_users , y=data_frame.greedy_user/ data_frame.num_users, label = "greedy")
sns.scatterplot(x=data_frame.num_users , y=data_frame.greedy_user / data_frame.num_users)

sns.lineplot(x=data_frame.num_users , y=data_frame.mcf_user/ data_frame.num_users, label = "most capacity first")
sns.scatterplot(x=data_frame.num_users , y=data_frame.mcf_user/ data_frame.num_users)

sns.lineplot(x=data_frame.num_users , y=data_frame.mflow_user/ data_frame.num_users, label = "constrained max flow")
sns.scatterplot(x=data_frame.num_users , y=data_frame.mflow_user/ data_frame.num_users)

plt.ylim(0,0.12)
plt.legend(fontsize=20)
plt.xlabel("Total number of users",fontsize=20)
plt.ylabel("fraction of users allocated",fontsize=20)
# plt.title("Fraction of users allocated with common deadline 10",fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.savefig("fraction_of_users_vs_total_users.eps" , format='eps')

plt.show()
```
