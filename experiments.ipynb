```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
### NUMBER OF SERVERS
tdf = pd.DataFrame()
tdf['NumServer']  = [12  , 20  , 26  , 29  , 33  , 35 ,  38 , 40 , 43 , 51 ]
tdf['Users']      = [443 , 869 , 870 , 872 , 874 , 874, 874 , 874, 874, 879]
tdf['GreedyU']    = [19  , 27  , 32  , 38  ,  40 ,  42,  44 , 45 ,  48, 58 ]
tdf['RandomU']    = [17  , 23  , 29  , 35  ,  37 ,  38,  40 , 41 ,  43, 53 ]
tdf['McfU']       = [19  , 29  , 36  , 46  ,  49 ,  51,  53 , 54 ,  61, 74 ]
tdf['MMcfU']      = [30  , 40  , 47  , 57  ,  59 ,  61,  63 , 64 ,  71, 87 ]
tdf['RoU']        = [15  , 21  , 25  , 31  ,  33 ,  34,  36 , 37 ,  39, 48 ]

tdf['GreedyS']    = [9  , 13   , 17  , 20  ,  22 ,  23,  25 , 26 ,  27, 33 ]
tdf['RandomS']    = [9  , 13  ,  17  , 20  ,  22 ,  23,  25 , 26 ,  27, 33 ]
tdf['McfS']       = [9  , 13  ,  17  , 20  ,  22 ,  23,  25 , 26 ,  27, 32 ]
tdf['MMcfS']      = [9  , 13  ,  17  , 20  ,  22 ,  23,  25 , 26 ,  27, 32 ]
tdf['RoS']        = [9  , 13  ,  17  , 20  ,  22 ,  23,  25 , 26 ,  27, 33 ]
xdf = tdf
```

## Number of users vs the number of servers

```python
## number of users allocated vs the number of total users
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style = 'whitegrid')
plt.figure(figsize = (15 , 10))
sns.lineplot(x = xdf.NumServer  , y = xdf.GreedyU)
sns.lineplot(x = xdf.NumServer  , y = xdf.RandomU)
sns.lineplot(x = xdf.NumServer  , y = xdf.McfU)
sns.lineplot(x = xdf.NumServer  , y = xdf.MMcfU)
sns.lineplot(x = xdf.NumServer  , y = xdf.RoU)
plt.legend(['Greedy' , 'Random' , 'MCF' , 'MMCF' , 'RO'])
plt.xlabel('Number of servers')
plt.ylabel('Number of users allocated')
plt.savefig('usersvsserver.eps' , format='eps')
plt.show()
```

### Number of actual servers vs allocated number of servers

```python
## number of users allocated vs the number of total users
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style = 'whitegrid')
plt.figure(figsize = (15 , 10))
sns.lineplot(x = xdf.NumServer  , y = xdf.GreedyS)
sns.lineplot(x = xdf.NumServer  , y = xdf.RandomS)
sns.lineplot(x = xdf.NumServer  , y = xdf.McfS)
sns.lineplot(x = xdf.NumServer  , y = xdf.MMcfS)
sns.lineplot(x = xdf.NumServer  , y = xdf.RoS)
plt.legend(['Greedy' , 'Random' , 'MCF' , 'MMCF' , 'RO'])
plt.xlabel('Number of servers')
plt.ylabel('Number of servers allocated')
plt.savefig('servervsservers.eps' , format='eps')
plt.show()
```

## Percentage of users allocated vs the number of servers

```python
## number of users allocated vs the number of total users
import seaborn as sns
import matplotlib.pyplot as plt
xdf['perGreedyU'] = xdf['GreedyU'] / xdf['Users'] 
xdf['perRandomU'] = xdf['RandomU'] / xdf['Users'] 
xdf['perMcfU']    = xdf['McfU'] / xdf['Users'] 
xdf['perMMcfU']   = xdf['MMcfU'] / xdf['Users'] 
xdf['perRoU']     = xdf['RoU'] / xdf['Users'] 

sns.set(style = 'whitegrid')
plt.figure(figsize = (15 , 10))
sns.lineplot(x = xdf.NumServer  , y = xdf.perGreedyU)
sns.lineplot(x = xdf.NumServer  , y = xdf.perRandomU)
sns.lineplot(x = xdf.NumServer  , y = xdf.perMcfU)
sns.lineplot(x = xdf.NumServer  , y = xdf.perMMcfU)
sns.lineplot(x = xdf.NumServer  , y = xdf.perRoU)
plt.legend(['Greedy' , 'Random' , 'MCF' , 'MMCF' , 'RO'])
plt.xlabel('Number of servers')
plt.ylabel('Percentage users allocated')
plt.savefig('percentageusersvsservers.eps' , format='eps')
plt.show()
```
