```python
import pandas as pd
import numpy as np
```

```python
eua = pd.read_csv ('eua_experiment.csv')
```

```python
eua.drop(columns=['Unnamed: 0'] , inplace = True)
```

```python
eua
```

## GENERATING RANDOM RESOURCES FOR USERS

```python
uf = eua.groupby(['userLat' , 'userLong']).count()[['coverage']].reset_index()
uf
```

```python
userResource = [[1,2,1,2],[2,3,3,4],[5,7,6,6]]
```

```python
from random import choice
userResourcedict = {}
for index in range(uf.shape[0]):
    latitude = uf.iloc[index]['userLat']
    longitude = uf.iloc[index]['userLong']
    usr = (latitude,  longitude)
    if (usr not in userResourcedict):
        userResourcedict[usr] = choice(userResource)
```

## GENERATING CAPACITIES FOR SERVERS

```python
sf = eua.groupby(['serverLat' , 'serverLong']).count().reset_index()
sf
```

```python
serverCPU = np.random.normal(loc=10,scale=10,size=sf.shape[0])
serverRAM = np.random.normal(loc=10,scale=10,size=sf.shape[0])
serverStorage = np.random.normal(loc=10,scale=10,size=sf.shape[0])
serverBandWidth = np.random.normal(loc=10,scale=10,size=sf.shape[0])

serverResourcedict = {}
for index in range(sf.shape[0]):
    latitude = sf.iloc[index]['serverLat']
    longitude = sf.iloc[index]['serverLong']
    server = (latitude,  longitude)
    CPU = round (max(1, serverCPU[index]))
    RAM = round (max(1, serverRAM[index]))
    STO = round (max(1 , serverStorage[index]))
    BAN = round (max (1, serverBandWidth[index]))
    serverResourcedict[server] = [CPU , RAM , STO, BAN]
serverResourcedict
```

### GENERATING THE FINAL DATASET

```python
serverCPU=[]
serverRAM=[]
serverSTO=[]
serverBAN=[]
userCPU=[]
userRAM=[]
userSTO=[]
userBAN=[]
for index in range(eua.shape[0]):
    if (index % 1000 == 0):
        print (index )
    serverLat = eua.iloc[index]['serverLat']
    serverLong = eua.iloc[index]['serverLong']
    userLat = eua.iloc[index]['userLat']
    userLong = eua.iloc[index]['userLong']
    server = (serverLat , serverLong)
    user = (userLat, userLong)
    serverCPU.append (serverResourcedict[server][0])
    serverRAM.append (serverResourcedict[server][1])
    serverSTO.append (serverResourcedict[server][2])
    serverBAN.append (serverResourcedict[server][3])
    userCPU.append (userResourcedict[user][0])
    userRAM.append (userResourcedict[user][1])
    userSTO.append (userResourcedict[user][2])
    userBAN.append (userResourcedict[user][3])
    
```

```python
eua['SCPU'] = serverCPU
eua['SRAM'] = serverRAM
eua['SSTO'] = serverSTO
eua['SBAN'] = serverBAN
eua['UCPU'] = userCPU
eua['URAM'] = userRAM
eua['USTO'] = userSTO
eua['UBAN'] = userBAN
```
