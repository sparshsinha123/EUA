```python
def getRadian (deg):
    return ((deg * np.pi) / 180.0)
def distanceBetweenPoints (latitude1 , longitude1 , latitude2 , longitude2):
    radiusOfEarth = 6378.1
    lat1 = getRadian(latitude1)
    lat2 = getRadian(latitude2)
    long1 = getRadian(longitude1)
    long2 = getRadian(longitude2)
    return radiusOfEarth * math.acos((math.sin(lat1)*math.sin(lat2))+math.cos(lat1)*math.cos(lat2)*math.cos(long2-long1))
def getArea (distance):
    return np.pi * distance * distance
```

## EXPERIMENTAL SETUP

```python
NUMBER_OF_SERVERS = 55
COVERAGE_LOW = 100
COVERAGE_HIGH = 150
```

```python
filtered_edges = edge_servers.head(NUMBER_OF_SERVERS)
filtered_edges = filtered_edges[['SITE_ID', 'LATITUDE' , 'LONGITUDE' , 'STATE' ]]
filtered_edges['COVERAGE'] = np.random.uniform(low = COVERAGE_LOW , high = COVERAGE_HIGH , size = NUMBER_OF_SERVERS)
filtered_edges
```

```python
numServers = filtered_edges.shape[0]
cnt = 0
edgeServerSiteId = []
uId = []
edgeServersLatList = []
edgeServersLongList = []
edgeServersStateList = []
edgeServersCoverageList = []
usersLatList = []
usersLongList = []

for index in range (numServers):
    if (index % 10 == 0):
        print(index)
    site = filtered_edges.iloc[index]['SITE_ID']
    latitude = filtered_edges.iloc[index]['LATITUDE']
    longitude = filtered_edges.iloc[index]['LONGITUDE']
    coverage = filtered_edges.iloc[index]['COVERAGE']
    numUsers = users.shape[0]
    for userId in range(numUsers):
        userLat = users.iloc[userId]['Latitude']
        userLong = users.iloc[userId]['Longitude']
        distance_eua = distanceBetweenPoints (latitude, longitude, userLat, userLong)
        if (distance_eua <= coverage):
            usersLatList.append(userLat)
            usersLongList.append(userLong)
            edgeServersLatList.append(latitude)
            edgeServersLongList.append(longitude)
            edgeServersCoverageList.append(coverage)
            edgeServerSiteId.append(site)
            uId.append (userId)
```

```python
eua_data_set = pd.DataFrame()
eua_data_set['siteId'] = edgeServerSiteId
eua_data_set['serverLat'] = edgeServersLatList
eua_data_set['serverLong'] = edgeServersLongList
eua_data_set['coverage'] = edgeServersCoverageList
eua_data_set['uId'] = uId;
eua_data_set['userLat'] = usersLatList
eua_data_set['userLong'] = usersLongList

```

```python
eua_data_set.to_csv('eua_experiment.csv')
```
